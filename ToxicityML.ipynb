## Train Random Forest Model

### Subtask:
Initialize and train a RandomForestClassifier using the `X_train` and `y_train` datasets. We will start with default parameters and then consider tuning if necessary.


```python
from sklearn.ensemble import RandomForestClassifier

# Initialize a RandomForestClassifier with default parameters
model = RandomForestClassifier(random_state=42)

# Train the model using the training data
model.fit(X_train, y_train)

print("RandomForestClassifier model trained successfully.")
```

## Evaluate Model Performance

### Subtask:
Evaluate the trained Random Forest model's performance on the `X_test` and `y_test` datasets. We'll look at metrics such as accuracy, precision, recall, and F1-score, and display a classification report and confusion matrix to understand its performance in detail. I will also generate plots to visualize the results.


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Make predictions on the test data
y_pred = model.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, pos_label='Toxic')
recall = recall_score(y_test, y_pred, pos_label='Toxic')
f1 = f1_score(y_test, y_pred, pos_label='Toxic')

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")

# Display classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=['NonToxic', 'Toxic'])

# Visualize confusion matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['NonToxic', 'Toxic'], yticklabels=['NonToxic', 'Toxic'])
plt.title('Confusion Matrix for Random Forest Classifier')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
